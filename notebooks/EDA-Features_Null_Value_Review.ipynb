# Calculate percentage of NULL value in each feature

```python
from utils.misc_utils import connect_rds
import pandas as pd 
conn = connect_rds()
```

```python
import csv
feat_list = list()
table_list = list()
with open('V3_Feature_list.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
            feature = row['name']
            if row['ty'] != "knmi" and row['ty'] != "temporal" :
                table_name = "features_"+row['ty']+"_agg."+row['table']
            else:
                table_name = row['table']
            feat_list.append(feature)
            table_list.append(table_name)
            
```

```python
null_perc_list = list()
for feat, table in zip(feat_list, table_list):
    sql = f"""
    with k as( select  COUNT(1)-COUNT({feat}) as null_count,
                       COUNT(1)::float as total_count
    from {table})
    select 100.0*null_count/total_count::float as perc, null_count, total_count from k;"""
    df = pd.read_sql(sql, con=conn)
    print(table,feat)
    print("Null Percentage:",df.perc.values[0])
    null_perc_list.append(df.perc.values[0])
```

```python
for null_perc in null_perc_list:
    print(null_perc)
```
